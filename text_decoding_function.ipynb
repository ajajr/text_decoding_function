```python
message = ".../coding_qual_input.txt"
```

```python
def create_staircase(nums):
  nums.sort()
  step = 1
  subsets = []
  while len(nums) != 0:
    if len(nums) >= step:
      subsets.append(nums[0:step])
      nums = nums[step:]
      step += 1
    else:
      return False
      
  return subsets

def decode(message_file):
    num_list = []
    word_dic = {}
    file = open(message_file, "r")
    for line in file.readlines():
        lis = line.strip("\n").split(" ")
        num_list.append(int(lis[0]))
        word_dic[lis[0]]=  lis[1]
    num_list.sort()
    pyramid = create_staircase(num_list)
    pyramid.sort()
    sentence = ""
    space = " "
    i = 0
    for i in range(len(pyramid)):
       if i == 0:
          sentence = word_dic[str(pyramid[i][-1])]
       else:
          sentence = sentence + space + word_dic[str(pyramid[i][-1])]
    return sentence
```

```python
decode(message)
```
